## 1. Querying MySQL

```python
import pymysql.cursors
```

```python
HOST = "localhost"
USER = "your_username"
DATABASE = "skaila"
PASSWORD = "your_password"
PORT = 3306
```

```python
conn = pymysql.connect(
    host=HOST, user=USER, port=PORT,
    password=PASSWORD, database=DATABASE)
conn
```

```python
cursor = conn.cursor()
```

#### 1.1 Pull all films with a Rental Rate of $0.99

Write a query to pull all films with a rental_rate of 0.99. Fetch only film_id, title, rental_rate and length. 

```python
cursor.execute("""SELECT film_id, title, rental_rate, length
                  FROM film
                  WHERE rental_rate = 0.99""")
films_under_a_buck = cursor.fetchall()
films_under_a_buck 
```

#### 1.2 Pull all the films

Now, let's take it one step back, pull ALL the films with their film_id, title, rental_rate and length. Store your results in the variable all_films.

```python
cursor.execute("""SELECT film_id, title, rental_rate, length
                  FROM film""")
all_films = cursor.fetchall()
all_films
```

#### 1.3 Calculate `price_per_minute`

Price per minute is defined as the rental rate of the film divided by its length. Use the data from all_films to create a dictionary named films_ppm that contains the title as the key and ppm as its value.

```python
from pymysql.cursors import DictCursor
cursor = conn.cursor(DictCursor)
```

```python
films_ppm = {}
for film in all_films:
    ppm = film['rental_rate'] / film['length']
    films_ppm[film['title']] = ppm
films_ppm
```
