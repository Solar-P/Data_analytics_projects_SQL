#### 1.3 Calculate `price_per_minute`

Price per minute is defined as the rental rate of the film divided by its length. Use the data from all_films to create a dictionary named films_ppm that contains the title as the key and ppm as its value.

```python
from pymysql.cursors import DictCursor
cursor = conn.cursor(DictCursor)
```

```python
films_ppm = {}
for film in all_films:
    ppm = film[2] / film[3]  # Accessing rental_rate and length using indices
    films_ppm[film[1]] = ppm  # Accessing title using index
```
